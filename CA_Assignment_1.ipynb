```python


# --- 1. INSTALL & IMPORTS ---
!pip install xgboost -q

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import GridSearchCV, cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')
plt.style.use('seaborn-v0_8-whitegrid')
print("All libraries imported successfully.\n")

# --- 2. LOAD DATASET ---
from google.colab import files
uploaded = files.upload()

df = pd.read_excel('PriceHistory__5_.xlsx.xlsx', header=15)
print("=" * 60)
print("DATASET OVERVIEW")
print("=" * 60)
print(f"Shape: {df.shape}")
print(f"Columns: {df.columns.tolist()}")
print(f"\nData Types:\n{df.dtypes}")
print(f"\nFirst 5 rows:\n{df.head()}")
print(f"\nMissing Values:\n{df.isnull().sum()}")
print(f"\nDescriptive Statistics:\n{df.describe()}")

# --- 3. DATA CLEANING ---
df_clean = df.dropna(subset=['Change', '% Change', '% Return']).copy()
df_clean = df_clean.sort_values('Date').reset_index(drop=True)
df_clean['Cumulative Return %'] = df_clean['Cumulative Return %'].fillna(0)
df_clean['Cumulative Change %'] = df_clean['Cumulative Change %'].fillna(0)
print(f"\nOriginal shape: {df.shape}")
print(f"Cleaned shape:  {df_clean.shape}")
print(f"Date range: {df_clean['Date'].min()} to {df_clean['Date'].max()}")

# --- 4. FEATURE ENGINEERING ---
# Simple Moving Averages
df_clean['SMA_5'] = df_clean['Price'].rolling(window=5).mean()
df_clean['SMA_10'] = df_clean['Price'].rolling(window=10).mean()
df_clean['SMA_20'] = df_clean['Price'].rolling(window=20).mean()
df_clean['SMA_50'] = df_clean['Price'].rolling(window=50).mean()

# Exponential Moving Averages
df_clean['EMA_12'] = df_clean['Price'].ewm(span=12, adjust=False).mean()
df_clean['EMA_26'] = df_clean['Price'].ewm(span=26, adjust=False).mean()

# MACD
df_clean['MACD'] = df_clean['EMA_12'] - df_clean['EMA_26']
df_clean['MACD_Signal'] = df_clean['MACD'].ewm(span=9, adjust=False).mean()

# RSI (14-day)
def compute_rsi(series, window=14):
    delta = series.diff()
    gain = delta.where(delta > 0, 0.0)
    loss = (-delta).where(delta < 0, 0.0)
    avg_gain = gain.rolling(window=window).mean()
    avg_loss = loss.rolling(window=window).mean()
    rs = avg_gain / avg_loss
    return 100 - (100 / (1 + rs))

df_clean['RSI_14'] = compute_rsi(df_clean['Price'], window=14)

# Bollinger Bands (20-day)
df_clean['BB_Middle'] = df_clean['Price'].rolling(window=20).mean()
bb_std = df_clean['Price'].rolling(window=20).std()
df_clean['BB_Upper'] = df_clean['BB_Middle'] + (2 * bb_std)
df_clean['BB_Lower'] = df_clean['BB_Middle'] - (2 * bb_std)
df_clean['BB_Width'] = df_clean['BB_Upper'] - df_clean['BB_Lower']

# Lag Features
df_clean['Price_Lag_1'] = df_clean['Price'].shift(1)
df_clean['Price_Lag_2'] = df_clean['Price'].shift(2)
df_clean['Price_Lag_3'] = df_clean['Price'].shift(3)
df_clean['Price_Lag_5'] = df_clean['Price'].shift(5)

# Additional Features
df_clean['Daily_Range'] = df_clean['High'] - df_clean['Low']
df_clean['Price_to_SMA20_Ratio'] = df_clean['Price'] / df_clean['SMA_20']
df_clean['Volume_Change'] = df_clean['CVol'].pct_change()

print(f"\nFeature Engineering Complete! New features created: {len(df_clean.columns) - len(df.columns)}")

# Drop NaN rows from rolling/lag
df_feat = df_clean.dropna().copy()
print(f"Shape after dropping NaN: {df_feat.shape}")

# --- 5. VISUALISE ENGINEERED FEATURES ---
fig, axes = plt.subplots(3, 1, figsize=(14, 12), sharex=True)

axes[0].plot(df_feat['Date'], df_feat['Price'], label='Price', linewidth=1)
axes[0].plot(df_feat['Date'], df_feat['SMA_20'], label='SMA 20', linewidth=1, alpha=0.8)
axes[0].plot(df_feat['Date'], df_feat['SMA_50'], label='SMA 50', linewidth=1, alpha=0.8)
axes[0].fill_between(df_feat['Date'], df_feat['BB_Lower'], df_feat['BB_Upper'],
                     alpha=0.15, color='grey', label='Bollinger Bands')
axes[0].set_title('Goldman Sachs — Price with Moving Averages & Bollinger Bands', fontsize=13)
axes[0].set_ylabel('Price ($)')
axes[0].legend(loc='upper left', fontsize=9)

axes[1].plot(df_feat['Date'], df_feat['RSI_14'], color='purple', linewidth=1)
axes[1].axhline(70, color='red', linestyle='--', alpha=0.7, label='Overbought (70)')
axes[1].axhline(30, color='green', linestyle='--', alpha=0.7, label='Oversold (30)')
axes[1].set_title('RSI (14-day)', fontsize=13)
axes[1].set_ylabel('RSI')
axes[1].legend(loc='upper left', fontsize=9)
axes[1].set_ylim(10, 90)

axes[2].plot(df_feat['Date'], df_feat['MACD'], label='MACD', linewidth=1)
axes[2].plot(df_feat['Date'], df_feat['MACD_Signal'], label='Signal Line', linewidth=1)
axes[2].bar(df_feat['Date'], df_feat['MACD'] - df_feat['MACD_Signal'],
            alpha=0.3, color='grey', label='MACD Histogram')
axes[2].set_title('MACD', fontsize=13)
axes[2].set_ylabel('MACD Value')
axes[2].legend(loc='upper left', fontsize=9)

plt.xlabel('Date')
plt.tight_layout()
plt.show()

# --- 6. TARGET VARIABLE & TRAIN/TEST SPLIT ---
df_model = df_feat.copy()
df_model['Target'] = df_model['Price'].shift(-1)
df_model = df_model.dropna(subset=['Target'])

feature_columns = [
    'Price_Lag_1', 'Price_Lag_2', 'Price_Lag_3', 'Price_Lag_5',
    'SMA_5', 'SMA_10', 'SMA_20', 'SMA_50',
    'EMA_12', 'EMA_26', 'RSI_14',
    'MACD', 'MACD_Signal', 'BB_Width',
    'Daily_Range', 'Price_to_SMA20_Ratio',
    'Volume_Change', 'CVol', 'Change', '% Change'
]

X = df_model[feature_columns]
y = df_model['Target']

split_index = int(len(X) * 0.80)
X_train, X_test = X.iloc[:split_index], X.iloc[split_index:]
y_train, y_test = y.iloc[:split_index], y.iloc[split_index:]

print("\n" + "=" * 60)
print("TRAIN / TEST SPLIT SUMMARY")
print("=" * 60)
print(f"Total samples:    {len(X)}")
print(f"Training samples: {len(X_train)} ({len(X_train)/len(X)*100:.1f}%)")
print(f"Testing samples:  {len(X_test)} ({len(X_test)/len(X)*100:.1f}%)")
print(f"Training period:  {df_model['Date'].iloc[:split_index].min().date()} → {df_model['Date'].iloc[:split_index].max().date()}")
print(f"Testing period:   {df_model['Date'].iloc[split_index:].min().date()} → {df_model['Date'].iloc[split_index:].max().date()}")
print(f"Number of features: {len(feature_columns)}")

# --- 7. SCALE FEATURES ---
scaler = StandardScaler()
X_train_scaled = pd.DataFrame(scaler.fit_transform(X_train), columns=feature_columns, index=X_train.index)
X_test_scaled = pd.DataFrame(scaler.transform(X_test), columns=feature_columns, index=X_test.index)
print("\nFeature scaling complete (StandardScaler).")

# --- 8. HELPER FUNCTIONS ---
def evaluate_model(model_name, y_true, y_pred, y_train_true=None, y_train_pred=None):
    metrics = {
        'Model': model_name,
        'MAE': mean_absolute_error(y_true, y_pred),
        'RMSE': np.sqrt(mean_squared_error(y_true, y_pred)),
        'R² Score': r2_score(y_true, y_pred),
        'MAPE (%)': np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    }
    if y_train_true is not None and y_train_pred is not None:
        metrics['Train R²'] = r2_score(y_train_true, y_train_pred)
    print("\n" + "=" * 60)
    print(f"  {model_name} — TEST SET PERFORMANCE")
    print("=" * 60)
    for key, val in metrics.items():
        if key != 'Model':
            print(f"  {key:20s}: {val:.6f}" if isinstance(val, float) else f"  {key}: {val}")
    print("=" * 60)
    return metrics

def plot_predictions(y_true, y_pred, title, dates=None):
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    if dates is not None:
        axes[0].plot(dates, y_true.values, label='Actual', linewidth=1)
        axes[0].plot(dates, y_pred, label='Predicted', linewidth=1, alpha=0.8)
        axes[0].set_xlabel('Date')
    else:
        axes[0].plot(y_true.values, label='Actual', linewidth=1)
        axes[0].plot(y_pred, label='Predicted', linewidth=1, alpha=0.8)
    axes[0].set_title(f'{title} — Actual vs Predicted')
    axes[0].set_ylabel('Price ($)')
    axes[0].legend()
    axes[1].scatter(y_true, y_pred, alpha=0.5, s=15, edgecolors='k', linewidth=0.3)
    min_val = min(y_true.min(), min(y_pred))
    max_val = max(y_true.max(), max(y_pred))
    axes[1].plot([min_val, max_val], [min_val, max_val], 'r--', linewidth=1.5, label='Perfect Prediction')
    axes[1].set_xlabel('Actual Price ($)')
    axes[1].set_ylabel('Predicted Price ($)')
    axes[1].set_title(f'{title} — Scatter Plot')
    axes[1].legend()
    plt.tight_layout()
    plt.show()

test_dates = df_model['Date'].iloc[split_index:].values

# ============================================================
# MODEL 1 — LINEAR / MULTIPLE REGRESSION
# ============================================================
print("\n\n" + "#" * 60)
print("# MODEL 1 — LINEAR / MULTIPLE REGRESSION")
print("#" * 60)

lr_model = LinearRegression()
lr_model.fit(X_train_scaled, y_train)
y_train_pred_lr = lr_model.predict(X_train_scaled)
y_pred_lr = lr_model.predict(X_test_scaled)
lr_metrics = evaluate_model('Linear Regression', y_test, y_pred_lr, y_train, y_train_pred_lr)

plot_predictions(y_test, y_pred_lr, 'Linear Regression', dates=test_dates)

# Feature Coefficients
lr_coefs = pd.DataFrame({'Feature': feature_columns, 'Coefficient': lr_model.coef_}).sort_values('Coefficient', ascending=True)
plt.figure(figsize=(10, 7))
plt.barh(lr_coefs['Feature'], lr_coefs['Coefficient'], color='steelblue')
plt.title('Linear Regression — Feature Coefficients (Standardised)', fontsize=13)
plt.xlabel('Coefficient Value')
plt.tight_layout()
plt.show()
print(f"Intercept: {lr_model.intercept_:.4f}")

# ============================================================
# MODEL 2 — DECISION TREE REGRESSOR
# ============================================================
print("\n\n" + "#" * 60)
print("# MODEL 2 — DECISION TREE REGRESSOR")
print("#" * 60)

dt_model = DecisionTreeRegressor(random_state=42, max_depth=10, min_samples_split=10, min_samples_leaf=5)
dt_model.fit(X_train_scaled, y_train)
y_train_pred_dt = dt_model.predict(X_train_scaled)
y_pred_dt = dt_model.predict(X_test_scaled)
dt_metrics = evaluate_model('Decision Tree', y_test, y_pred_dt, y_train, y_train_pred_dt)

plot_predictions(y_test, y_pred_dt, 'Decision Tree', dates=test_dates)

# Hyperparameter Tuning — Decision Tree
print("\nTuning Decision Tree hyperparameters with GridSearchCV...")
dt_param_grid = {
    'max_depth': [5, 8, 10, 15, 20],
    'min_samples_split': [5, 10, 20, 30],
    'min_samples_leaf': [3, 5, 10, 15],
    'max_features': ['sqrt', 'log2', None]
}
dt_grid_search = GridSearchCV(DecisionTreeRegressor(random_state=42), param_grid=dt_param_grid,
                               cv=5, scoring='r2', n_jobs=-1, verbose=0)
dt_grid_search.fit(X_train_scaled, y_train)
print(f"Best Parameters: {dt_grid_search.best_params_}")
print(f"Best CV R² Score: {dt_grid_search.best_score_:.6f}")

dt_tuned = dt_grid_search.best_estimator_
y_train_pred_dt_tuned = dt_tuned.predict(X_train_scaled)
y_pred_dt_tuned = dt_tuned.predict(X_test_scaled)
dt_tuned_metrics = evaluate_model('Decision Tree (Tuned)', y_test, y_pred_dt_tuned, y_train, y_train_pred_dt_tuned)

plot_predictions(y_test, y_pred_dt_tuned, 'Decision Tree (Tuned)', dates=test_dates)

# Decision Tree Feature Importance
dt_importance = pd.DataFrame({'Feature': feature_columns, 'Importance': dt_tuned.feature_importances_}).sort_values('Importance', ascending=True)
plt.figure(figsize=(10, 7))
plt.barh(dt_importance['Feature'], dt_importance['Importance'], color='darkorange')
plt.title('Decision Tree (Tuned) — Feature Importance', fontsize=13)
plt.xlabel('Importance')
plt.tight_layout()
plt.show()

# ============================================================
# MODEL 3 — RANDOM FOREST REGRESSOR
# ============================================================
print("\n\n" + "#" * 60)
print("# MODEL 3 — RANDOM FOREST REGRESSOR")
print("#" * 60)

rf_model = RandomForestRegressor(n_estimators=100, random_state=42, max_depth=15,
                                  min_samples_split=5, min_samples_leaf=3, n_jobs=-1)
rf_model.fit(X_train_scaled, y_train)
y_train_pred_rf = rf_model.predict(X_train_scaled)
y_pred_rf = rf_model.predict(X_test_scaled)
rf_metrics = evaluate_model('Random Forest', y_test, y_pred_rf, y_train, y_train_pred_rf)

plot_predictions(y_test, y_pred_rf, 'Random Forest', dates=test_dates)

# Hyperparameter Tuning — Random Forest
print("\nTuning Random Forest hyperparameters with GridSearchCV (this may take 2-3 minutes)...")
rf_param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 15, 20, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 3, 5]
}
rf_grid_search = GridSearchCV(RandomForestRegressor(random_state=42, n_jobs=-1),
                               param_grid=rf_param_grid, cv=3, scoring='r2', n_jobs=-1, verbose=1)
rf_grid_search.fit(X_train_scaled, y_train)
print(f"Best Parameters: {rf_grid_search.best_params_}")
print(f"Best CV R² Score: {rf_grid_search.best_score_:.6f}")

rf_tuned = rf_grid_search.best_estimator_
y_train_pred_rf_tuned = rf_tuned.predict(X_train_scaled)
y_pred_rf_tuned = rf_tuned.predict(X_test_scaled)
rf_tuned_metrics = evaluate_model('Random Forest (Tuned)', y_test, y_pred_rf_tuned, y_train, y_train_pred_rf_tuned)

plot_predictions(y_test, y_pred_rf_tuned, 'Random Forest (Tuned)', dates=test_dates)

# Random Forest Feature Importance
rf_importance = pd.DataFrame({'Feature': feature_columns, 'Importance': rf_tuned.feature_importances_}).sort_values('Importance', ascending=True)
plt.figure(figsize=(10, 7))
plt.barh(rf_importance['Feature'], rf_importance['Importance'], color='forestgreen')
plt.title('Random Forest (Tuned) — Feature Importance', fontsize=13)
plt.xlabel('Importance')
plt.tight_layout()
plt.show()

# ============================================================
# MODEL 4 — KNN (K-NEAREST NEIGHBORS)
# ============================================================
print("\n\n" + "#" * 60)
print("# MODEL 4 — KNN (K-NEAREST NEIGHBORS)")
print("#" * 60)

knn_model = KNeighborsRegressor(n_neighbors=5)
knn_model.fit(X_train_scaled, y_train)
y_train_pred_knn = knn_model.predict(X_train_scaled)
y_pred_knn = knn_model.predict(X_test_scaled)
knn_metrics = evaluate_model('KNN', y_test, y_pred_knn, y_train, y_train_pred_knn)

plot_predictions(y_test, y_pred_knn, 'KNN (K-Nearest Neighbors)', dates=test_dates)

# Hyperparameter Tuning — KNN
print("\nTuning KNN hyperparameters with GridSearchCV...")
knn_param_grid = {
    'n_neighbors': [3, 5, 7, 10, 15, 20],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan']
}
knn_grid_search = GridSearchCV(KNeighborsRegressor(), param_grid=knn_param_grid,
                                cv=5, scoring='r2', n_jobs=-1, verbose=0)
knn_grid_search.fit(X_train_scaled, y_train)
print(f"Best Parameters: {knn_grid_search.best_params_}")
print(f"Best CV R² Score: {knn_grid_search.best_score_:.6f}")

knn_tuned = knn_grid_search.best_estimator_
y_train_pred_knn_tuned = knn_tuned.predict(X_train_scaled)
y_pred_knn_tuned = knn_tuned.predict(X_test_scaled)
knn_tuned_metrics = evaluate_model('KNN (Tuned)', y_test, y_pred_knn_tuned, y_train, y_train_pred_knn_tuned)

plot_predictions(y_test, y_pred_knn_tuned, 'KNN (Tuned)', dates=test_dates)

# ============================================================
# MODEL 5 — XGBOOST
# ============================================================
print("\n\n" + "#" * 60)
print("# MODEL 5 — XGBOOST")
print("#" * 60)

xgb_model = XGBRegressor(n_estimators=100, max_depth=6, learning_rate=0.1,
                          random_state=42, n_jobs=-1, verbosity=0)
xgb_model.fit(X_train_scaled, y_train)
y_train_pred_xgb = xgb_model.predict(X_train_scaled)
y_pred_xgb = xgb_model.predict(X_test_scaled)
xgb_metrics = evaluate_model('XGBoost', y_test, y_pred_xgb, y_train, y_train_pred_xgb)

plot_predictions(y_test, y_pred_xgb, 'XGBoost', dates=test_dates)

# Hyperparameter Tuning — XGBoost
print("\nTuning XGBoost hyperparameters with GridSearchCV (this may take 2-3 minutes)...")
xgb_param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [4, 6, 8, 10],
    'learning_rate': [0.01, 0.05, 0.1],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0]
}
xgb_grid_search = GridSearchCV(XGBRegressor(random_state=42, n_jobs=-1, verbosity=0),
                                param_grid=xgb_param_grid, cv=3, scoring='r2', n_jobs=-1, verbose=1)
xgb_grid_search.fit(X_train_scaled, y_train)
print(f"Best Parameters: {xgb_grid_search.best_params_}")
print(f"Best CV R² Score: {xgb_grid_search.best_score_:.6f}")

xgb_tuned = xgb_grid_search.best_estimator_
y_train_pred_xgb_tuned = xgb_tuned.predict(X_train_scaled)
y_pred_xgb_tuned = xgb_tuned.predict(X_test_scaled)
xgb_tuned_metrics = evaluate_model('XGBoost (Tuned)', y_test, y_pred_xgb_tuned, y_train, y_train_pred_xgb_tuned)

plot_predictions(y_test, y_pred_xgb_tuned, 'XGBoost (Tuned)', dates=test_dates)

# XGBoost Feature Importance
xgb_importance = pd.DataFrame({'Feature': feature_columns, 'Importance': xgb_tuned.feature_importances_}).sort_values('Importance', ascending=True)
plt.figure(figsize=(10, 7))
plt.barh(xgb_importance['Feature'], xgb_importance['Importance'], color='crimson')
plt.title('XGBoost (Tuned) — Feature Importance', fontsize=13)
plt.xlabel('Importance')
plt.tight_layout()
plt.show()

# ============================================================
# MODEL COMPARISON SUMMARY — ALL 5 MODELS
# ============================================================
print("\n\n" + "#" * 60)
print("# MODEL COMPARISON SUMMARY — ALL 5 MODELS")
print("#" * 60)

results = pd.DataFrame([
    lr_metrics, dt_metrics, dt_tuned_metrics,
    rf_metrics, rf_tuned_metrics,
    knn_metrics, knn_tuned_metrics,
    xgb_metrics, xgb_tuned_metrics
])

print("\n" + "=" * 90)
print("                         MODEL COMPARISON — TEST SET PERFORMANCE")
print("=" * 90)
print(results[['Model', 'MAE', 'RMSE', 'R² Score', 'MAPE (%)', 'Train R²']].to_string(index=False))
print("=" * 90)

# Visual Comparison — Tuned Models Only
tuned_results = pd.DataFrame([lr_metrics, dt_tuned_metrics, rf_tuned_metrics, knn_tuned_metrics, xgb_tuned_metrics])
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
model_names = tuned_results['Model'].values
colors = ['steelblue', 'darkorange', 'forestgreen', 'purple', 'crimson']

axes[0].barh(model_names, tuned_results['R² Score'], color=colors)
axes[0].set_title('R² Score (higher is better)', fontsize=12)
axes[0].set_xlim(0, 1.05)
for i, v in enumerate(tuned_results['R² Score']):
    axes[0].text(v + 0.01, i, f'{v:.4f}', va='center', fontsize=9)

axes[1].barh(model_names, tuned_results['RMSE'], color=colors)
axes[1].set_title('RMSE (lower is better)', fontsize=12)
for i, v in enumerate(tuned_results['RMSE']):
    axes[1].text(v + 0.1, i, f'{v:.2f}', va='center', fontsize=9)

axes[2].barh(model_names, tuned_results['MAPE (%)'], color=colors)
axes[2].set_title('MAPE % (lower is better)', fontsize=12)
for i, v in enumerate(tuned_results['MAPE (%)']):
    axes[2].text(v + 0.01, i, f'{v:.2f}%', va='center', fontsize=9)

plt.suptitle('Model Comparison — Goldman Sachs Price Prediction (All 5 Models)', fontsize=14, y=1.02)
plt.tight_layout()
plt.show()

# Overlay All Tuned Predictions
plt.figure(figsize=(16, 6))
plt.plot(test_dates, y_test.values, label='Actual Price', linewidth=2, color='black')
plt.plot(test_dates, y_pred_lr, label='Linear Regression', linewidth=1, alpha=0.8)
plt.plot(test_dates, y_pred_dt_tuned, label='Decision Tree (Tuned)', linewidth=1, alpha=0.8)
plt.plot(test_dates, y_pred_rf_tuned, label='Random Forest (Tuned)', linewidth=1, alpha=0.8)
plt.plot(test_dates, y_pred_knn_tuned, label='KNN (Tuned)', linewidth=1, alpha=0.8)
plt.plot(test_dates, y_pred_xgb_tuned, label='XGBoost (Tuned)', linewidth=1, alpha=0.8)
plt.title('All 5 Models — Actual vs Predicted (Test Set)', fontsize=14)
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.legend(fontsize=9)
plt.tight_layout()
plt.show()

# ============================================================
# CROSS-VALIDATION — ALL TUNED MODELS
# ============================================================
print("\n\n" + "#" * 60)
print("# 5-FOLD CROSS-VALIDATION")
print("#" * 60)

cv_results = {}
models_cv = {
    'Linear Regression': LinearRegression(),
    'Decision Tree (Tuned)': dt_tuned,
    'Random Forest (Tuned)': rf_tuned,
    'KNN (Tuned)': knn_tuned,
    'XGBoost (Tuned)': xgb_tuned
}
for name, model in models_cv.items():
    cv_scores = cross_val_score(model, X_train_scaled, y_train, cv=5, scoring='r2', n_jobs=-1)
    cv_results[name] = cv_scores
    print(f"{name:30s} | CV R²: {cv_scores.mean():.6f} ± {cv_scores.std():.6f}")

plt.figure(figsize=(12, 5))
plt.boxplot(cv_results.values(), labels=cv_results.keys(), patch_artist=True,
            boxprops=dict(facecolor='lightblue'))
plt.title('5-Fold Cross-Validation R² Scores — All Models', fontsize=13)
plt.ylabel('R² Score')
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()

# ============================================================
# RESIDUAL ANALYSIS — BEST MODEL
# ============================================================
print("\n\n" + "#" * 60)
print("# RESIDUAL ANALYSIS — BEST MODEL")
print("#" * 60)

best_idx = tuned_results['R² Score'].idxmax()
best_model_name = tuned_results.loc[best_idx, 'Model']
print(f"Best performing model: {best_model_name}")

preds_map = {
    'Linear Regression': y_pred_lr,
    'Decision Tree (Tuned)': y_pred_dt_tuned,
    'Random Forest (Tuned)': y_pred_rf_tuned,
    'KNN (Tuned)': y_pred_knn_tuned,
    'XGBoost (Tuned)': y_pred_xgb_tuned
}
best_preds = preds_map[best_model_name]
residuals = y_test.values - best_preds

fig, axes = plt.subplots(1, 3, figsize=(16, 4))

axes[0].hist(residuals, bins=40, color='steelblue', edgecolor='black', alpha=0.7)
axes[0].axvline(0, color='red', linestyle='--')
axes[0].set_title(f'{best_model_name} — Residual Distribution')
axes[0].set_xlabel('Residual ($)')
axes[0].set_ylabel('Frequency')

axes[1].plot(test_dates, residuals, linewidth=0.8, alpha=0.8)
axes[1].axhline(0, color='red', linestyle='--')
axes[1].set_title('Residuals Over Time')
axes[1].set_xlabel('Date')
axes[1].set_ylabel('Residual ($)')

axes[2].scatter(best_preds, residuals, alpha=0.4, s=10)
axes[2].axhline(0, color='red', linestyle='--')
axes[2].set_title('Residuals vs Predicted')
axes[2].set_xlabel('Predicted Price ($)')
axes[2].set_ylabel('Residual ($)')

plt.suptitle(f'Residual Analysis — {best_model_name}', fontsize=14, y=1.02)
plt.tight_layout()
plt.show()

print(f"\nResidual Statistics:")
print(f"  Mean: {residuals.mean():.4f}")
print(f"  Std:  {residuals.std():.4f}")
print(f"  Min:  {residuals.min():.4f}")
print(f"  Max:  {residuals.max():.4f}")

print("\n\n" + "=" * 60)

print("=" * 60)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from google.colab import files
uploaded = files.upload()
df = pd.read_csv('PriceHistory (35)(Price History).csv')
print("Initial Info:")
print(df.info())
df = pd.read_csv('PriceHistory (35)(Price History).csv', header=2)
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%y')
df = df.sort_values('Date')
print("\nMissing Values:")
print(df.isnull().sum())
print("\nSummary Statistics:")
print(df.describe())

numeric_cols = df.select_dtypes(include=np.number).columns
corr = df[numeric_cols].corr()
plt.figure()
plt.plot(df['Date'], df['Price'])
plt.title("Goldman Sachs Price Trend")
plt.xlabel("Date")
plt.ylabel("Price")
plt.xticks(rotation=45)
plt.show()
plt.figure()
plt.plot(df['Date'], df['CVol'])
plt.title("Cumulative Volume Trend")
plt.xlabel("Date")
plt.ylabel("Cumulative Volume")
plt.xticks(rotation=45)
plt.show()
df['Daily_Return'] = df['Price'].pct_change()
plt.figure()
plt.hist(df['Daily_Return'], bins=50)
plt.title("Distribution of Daily Returns")
plt.xlabel("Returns")
plt.ylabel("Frequency")
plt.show()
plt.figure()
plt.imshow(corr)
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation Heatmap")
plt.colorbar()
plt.show()
df['MA_5'] = df['Price'].rolling(window=5).mean()
df['MA_20'] = df['Price'].rolling(window=20).mean()

plt.figure()
plt.plot(df['Date'], df['Price'], label='Price')
plt.plot(df['Date'], df['MA_5'], label='MA 5')
plt.plot(df['Date'], df['MA_20'], label='MA 20')
plt.legend()
plt.title("Price with Moving Averages")
plt.xticks(rotation=45)
plt.show()
def calculate_rsi(data, window=14):
    diff = data.diff(1).dropna()
    gain = diff.where(diff > 0, 0)
    loss = -diff.where(diff < 0, 0)
    avg_gain = gain.ewm(com=window-1, min_periods=window).mean()
    avg_loss = loss.ewm(com=window-1, min_periods=window).mean()
    rs = avg_gain / avg_loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

df['RSI_14'] = calculate_rsi(df['Price'])

plt.figure()
plt.plot(df['Date'], df['RSI_14'])
plt.title("RSI Indicator")
plt.xticks(rotation=45)
plt.show()
```
